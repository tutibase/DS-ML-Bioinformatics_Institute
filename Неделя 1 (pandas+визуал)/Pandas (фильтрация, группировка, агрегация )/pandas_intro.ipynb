# Группировка и агрегация (groupby в связке с aggregate)

```python
# рассчет среднего по полу

# as_index - использование данного столбца в качестве индекса
# aggregate - расчет средних через агрегацию
students_performance.groupby('gender', as_index = False) \
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns = {'math_score' : 'mean_math_score', 'reading_score' : 'mean_reading_score'})
```

```python
# пересечение используемых лейблов (получаем составной индекс)
mean_scores = students_performance.groupby(['gender', 'race/ethnicity']) \
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns = {'math_score' : 'mean_math_score', 'reading_score' : 'mean_reading_score'})

mean_scores
```

```python
# обращение по составному индексу
# mean_scores.loc[('female', 'group D')]

# срез по составному индексу
mean_scores.loc[[('female', 'group D'), ('female', 'group E')]]
```

```python
# такое же пересечение, но без составного индекса
students_performance.groupby(['gender', 'race/ethnicity'], as_index = False) \
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns = {'math_score' : 'mean_math_score', 'reading_score' : 'mean_reading_score'})
```

```python
# количество уникальных значений в кажом пересечении
students_performance.groupby(['gender', 'race/ethnicity']).math_score.nunique()
```

```python
# сортировка по полу и оценкам по математике, 
# ascending=False - поставить по убыванию
# потом групировка по полу
students_performance.sort_values(['gender', 'math_score'], ascending=False) \
    .groupby('gender').head(5)
```

```python
# добавление новой колонки
students_performance['total_score'] = students_performance.math_score + \
                                      students_performance.reading_score + \
                                      students_performance.writing_score

students_performance.head(3)
```

```python
# добавление новой колонки другим способом
# добавляем колонку с логарифмом от другой колонки
students_performance = students_performance.assign(total_score_log = np.log(students_performance.total_score))
students_performance.head(3)
```

```python
# удаление колонки
# axis = 1 - по колонкам, а не строкам
students_performance.drop(['total_score', 'lunch'], axis = 1)
```

```python
# algae = pd.read_csv('algae.csv')
algae = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
algae.head()
```

```python
# расчет средней концентрации веществ в каждом из родов
algae.groupby('genus').mean().head()
```

```python
# максимальное значение концентрации вещества для данной группы
algae.query('genus == "Fucus"')['alanin'].max()
```

```python
# минимальное, среднее и максимальное значения концентрации вещества для данной группы
algae.groupby('genus').aggregate(['min', 'mean', 'max']).loc['Fucus', 'alanin'].round(2)
```

```python
# Размах (максимум - минимум) сахарозы в бурых, зелёных и красных водоросля
algae.groupby('group')['sucrose'].max() - algae.groupby('group')['sucrose'].min()
```
